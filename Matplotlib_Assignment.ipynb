```python
def test(x):
    return len(x.strip())
```

```python
df['movie_title_len'] = df.movie_title.apply(test)
```

```python
USA=df[df.country=='USA']
UK=df[df.country=='UK']
Canada=df[df.country=='Canada']
```

```python
plt.scatter(USA.gross, USA.duration,
            color = 'red', label ='setosa')



plt.scatter(UK.gross, UK.duration,
            color = 'yellow', label ='UK')



plt.scatter(Canada.gross, Canada.duration,
            color = 'blue', label ='Canada')

plt.legend()
plt.xlabel('gross')
plt.ylabel('duration')
plt.title('gross vs duration')
plt.show()
```
